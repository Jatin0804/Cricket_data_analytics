```python
import pandas as pd
import json
```

Match results json file

```python
with open('dataset/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
data
```

```python
len(data)
```

```python
data[0][("matchSummary")]
```

```python
df_match = pd.DataFrame(data[0]["matchSummary"])
```

```python
df_match.head()
```

```python
df_match.tail()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row["team1"] + ' Vs ' + row['team2']
    key2 = row["team2"] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
match_ids_dict
```

```python
df_match.to_csv('new_data/match_results.csv', index = False)
```

Batting Summary json file

```python
with open('dataset/t20_wc_batting_summary.json') as f:
    data = json.load(f)
```

```python
data
```

```python
for record in data:
    print(record['battingSummary'])
```

```python
all_records = []
for record in data:
    all_records.extend(record['battingSummary'])
    
all_records
```

```python
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting.head(10)
```

```python
df_batting.tail(10)
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if pd.notna(x) else "not_out")
df_batting.head(15)
```

```python
df_batting.tail(15)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace("â€", ""))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace("\xa0", ""))
df_batting.head()
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
```

```python
df_batting.head()
```

```python
df_batting.to_csv('new_data/batting_summary.csv', index=False)
```

```python
df_batting.shape
```
